```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.head()
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchased_amount'})
```

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
```

```python
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis=1)
```

```python
df.columns
```

```python
age_revenue = df.groupby('age_group')['purchased_amount'].sum().sort_values()

plt.figure(figsize=(8,5))
age_revenue.plot(kind='bar')
plt.title('Total Revenue by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Purchased Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
gender_revenue = df.groupby('gender')['purchased_amount'].sum()

plt.figure(figsize=(6,4))
gender_revenue.plot(kind='bar')
plt.title('Gender-wise Total Spending')
plt.xlabel('Gender')
plt.ylabel('Total Purchased Amount')
plt.show()
```

```python
category_revenue = df.groupby('category')['purchased_amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
category_revenue.plot(kind='bar')
plt.title('Revenue by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Purchased Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
top_items = df['item_purchased'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_items.plot(kind='bar')
plt.title('Top 10 Most Purchased Items')
plt.xlabel('Item')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.show()
```

```python
discount_impact = df.groupby('discount_applied')['purchased_amount'].mean()

plt.figure(figsize=(6,4))
discount_impact.plot(kind='bar')
plt.title('Average Purchase Amount: Discount vs No Discount')
plt.xlabel('Discount Applied')
plt.ylabel('Average Purchased Amount')
plt.show()
```

```python
subscription_revenue = df.groupby('subscription_status')['purchased_amount'].mean()

plt.figure(figsize=(6,4))
subscription_revenue.plot(kind='bar')
plt.title('Average Spending by Subscription Status')
plt.xlabel('Subscription Status')
plt.ylabel('Average Purchased Amount')
plt.show()
```

```python
season_sales = df.groupby('season')['purchased_amount'].sum()

plt.figure(figsize=(7,4))
season_sales.plot(kind='bar')
plt.title('Season-wise Total Sales')
plt.xlabel('Season')
plt.ylabel('Total Purchased Amount')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(
    x='frequency_of_purchases',
    y='purchased_amount',
    data=df
)
plt.title('Purchase Frequency vs Spending')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Purchased Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.histplot(
    x='previous_purchases',
    y='purchased_amount',
    data=df
)
plt.title('Previous Purchases vs Purchased Amount')
plt.xlabel('Previous Purchases')
plt.ylabel('Purchased Amount')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(
    x='review_rating',
    y='purchased_amount',
    data=df
)
plt.title('Review Rating vs Purchased Amount')
plt.xlabel('Review Rating')
plt.ylabel('Purchased Amount')
plt.show()
```
